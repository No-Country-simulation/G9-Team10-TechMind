# TechMind - Búsqueda Semántica con TF-IDF

Este notebook genera exclusivamente el modelo de Búsqueda Semántica (Fase 4), utilizando los datos limpios de la Fase 2.
La clasificación supervisada ha sido delegada a Gemini en la Fase 5.

```python
import pandas as pd
import numpy as np
import joblib
import os
from sklearn.feature_extraction.text import TfidfVectorizer
import warnings
warnings.filterwarnings('ignore')

print("✅ Entorno inicializado.")
```

```python
# 1. Cargar Datos Limpios
DATASET_PATH = r'C:\Users\Maxi.DESKTOP-8LJC287\Desktop\Entornos_Antigravity\W5(Varios)\techmind\ai_engine\dataset_techmind_mvp.csv'
df = pd.read_csv(DATASET_PATH)
df['clean_text'] = df['clean_text'].fillna("")

print(f"✅ Dataset cargado. Total de documentos: {df.shape[0]}")
```

```python
# 2. Vectorización TF-IDF
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english', ngram_range=(1, 2))
matriz_tfidf_global = vectorizer.fit_transform(df['clean_text'])

print(f"✅ Matriz TF-IDF generada: {matriz_tfidf_global.shape}")
```

```python
# 3. Exportar Modelos para la API
OUTPUT_DIR = r'C:\Users\Maxi.DESKTOP-8LJC287\Desktop\Entornos_Antigravity\W5(Varios)\techmind\Repo\Data\5.API_Final\models'
os.makedirs(OUTPUT_DIR, exist_ok=True)

joblib.dump(vectorizer, os.path.join(OUTPUT_DIR, 'tfidf_vectorizer.joblib'))
joblib.dump(matriz_tfidf_global, os.path.join(OUTPUT_DIR, 'tfidf_matrix.joblib'))
joblib.dump(df[['title', 'clean_text']], os.path.join(OUTPUT_DIR, 'dataset_reference.joblib'))

print("✅ Pipeline persistido en directorio models.")
```
